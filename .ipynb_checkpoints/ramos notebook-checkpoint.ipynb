```python
df= pd.read_csv('ramos2019-month.txt',skiprows= 95, sep= '\t')
df.head()
```

```python
#if you just want the labels of columns df.columns it gives you a list of the names of your columns
#gets rid of unwanted columns from before
df_new = df.drop(["d18O-P.loba","Unnamed: 3"], axis=1)
df_new.head()
```

```python
#setting variables and equations for columns 
year_rounded= np.round(df_new['age_AD'],2)
year= np.floor(year_rounded)
Month= np.ceil(12*(df_new['age_AD']-year))
```

```python
df_new['Year']= year
df_new['Month']= Month
df_new.head()
```

```python
#values in linear regression equation 
c_value= 10.790
m_value= -0.068
```

```python
#getting our SST value

def fun (x):
    c_value= 10.790
    m_value= -0.068
    sst= (x-c_value)/m_value
    return sst
```

```python
fun(df_new['Sr/Ca-P.loba'])
```

```python
#adding sst to df with function
df_new['SST'] = fun(df_new['Sr/Ca-P.loba'])
df_new.head()
```

```python
#indexing 
df_index= df_new.set_index(['Year','Month'])
df_index
```

```python
#getting rid of months and getting the yearly average
df_avg= df_index.mean(level=0)
df_avg.head()
```

```python
#playing around with levels 
df_index.mean(level=1)
```
