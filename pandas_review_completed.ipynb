```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

#### Import the ENSO climate index

```python
os.chdir('/Users/brownscholar/Dropbox/BridgeUP_ClimateCoders/Data')

ENSO = pd.read_excel('NOAA_ENSOIndex.xlsx')
```

```python
ENSO = ENSO.set_index('Year')
len(ENSO)
```

#### Find the monthly mean

Remember!
1. "your data".mean(axis = **0**) - mean along the columns
1. "your data".mean(axis = **1**) - mean along the rows

```python
ENSO_mn = ENSO.mean(axis = 0)
```

#### Plot your result

```python
ENSO_mn.plot()

plt.figure()
plt.plot(ENSO_mn)
plt.show()
```

#### When is ENSO the strongest? Sort your values to check! 

```python
ENSO_mn.sort_values(ascending = False)
```
